### 여러 개의 엑셀 파일에서 데이터 수정하기

```python
import re

file_name = 'C:/myPycode/data/담당자별_판매량_Andy사원.xlsx'

new_file_name = re.sub('.xlsx', '2.xlsx', file_name)
new_file_name
```

```python
import glob
import re
import pandas as pd

# 원하는 문자열이 포함된 파일을 검색해 리스트를 할당한다.
excel_data_files1 = glob.glob('C:/myPycode/data/담당자별_판매량_*사원.xlsx')

# 리스트에 있는 엑셀 파일만큼 반복 수행한다.
for f in excel_data_files1:
    # 엑세 파일에서 DataFrame 형식으로 데이터를 가져온다.
    df = pd.read_excel(f)
    
    # 특정 열의 값을 변경한다..
    if(df.loc[1,'담당자']=='A'):
        df['담당자'] = 'Andy'
    elif (df. loc[1, '담당자'] == 'B'):
        df['담당자'] = 'Becky'
    elif (df. loc[1, '담당자'] == 'C'):
        df['담당자'] = 'Chris'
        
    # 엑셀 파일 이름에서 지정된 문자열 패턴을 찾아서 파일명을 변경한다.
    f_new = re.sub('.xlsx', '2.xlsx', f)
    print(f_new)
    
    # 수정되 데이터를 새로운 이름의 엑셀 파일로 저장한다.
    new_excel_file = pd.ExcelWriter(f_new, engine = 'xlsxwriter')
    df.to_excel(new_excel_file, index = False)
    new_excel_file.save()
```

### 엘셀 데이터 계산하기

#### 행 데이터의 합계 구하기

```python
import pandas as pd

df = pd.read_excel('C:/myPycode/data/담당자별_판매량_통합.xlsx')

hnadbag = df[(df['제품명']=='핸드백')]
handbag
```

```python
handbag.sum(axis=1)
```

```python
handbag_sum = pd.DataFrame(handbag.sum(axis = 1), columns = ['연간판매량'])
handbag_sum
```

```python
handbag_total = handbag.join(handbag_sum)  #handbag DataFrame에 hhandbag_sum의 데이터를 추가
handbag_total
```

```python
handbag_total.sort_values(by='연간판매량', ascending = False) # ascending 값 조절로 오름, 내림차순 결정
```

#### 열 데이터의 합계 구하기

```python
handbag_total.sum()
```

```python
handbag_sum2 = pd.DataFrame(handbag_total.sum(), columns = ['합계'])
handbag_sum2
```

```python
handbag_total2 = handbag_total.append(handbag_sum2.T)
handbag_total2.loc['합계', '제품명'] = '핸드백'   #'합계'행의 '제품명' 열의 값을 '핸드백'으로 변경
handbag_total2.loc['합계', '담당자'] = '전체'
handbag_total2.loc['합계', '지역'] = '전체'
handbag_total2
```
