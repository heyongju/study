```python
from bs4 import BeautifulSoup as bs
import openpyxl
from selenium import webdriver

driver = webdriver.Chrome('C:/chromedriver.exe')
url = 'https://www.melon.com/chart/'
driver.get(url)
```

```python
import openpyxl
import os

wb = openpyxl.Workbook()
sheet = wb.active
sheet.append(['앨범아트', '제목', '가수'])

html = driver.page_source
soup = bs(html,'lxml')

music_title = soup.select('td div.ellipsis.rank01 a')
music_titles = [title.get_text() for title in music_title]
music_artist = soup.select('td div.ellipsis.rank02 span')
music_artists = [artist.get_text() for artist in music_artist]


for k in range(len(music_artists)):
    sheet.append([k+1, music_titles[k].strip(), music_artists[k].strip()])
wb.save('c:/myPycode/melona.xlsx')
```

```python
import requests
import openpyxl

wb = openpyxl.Workbook()
ws = wb.worksheets[0]



image = soup.select('td div a img')
for k in range(len(music_artists)):

    
    image_url = image[k].get('src').replace('/melon/resize/120/quality/80/optimize', '')
    get_album = requests.get(image_url)
    folder = 'C:/myPycode/download'

    imageFile = open(os.path.join(folder, os.path.basename(image_url)),'wb')
    chunk_size = 1000000
    for chunk in get_album.iter_content(chunk_size):
        imageFile.write(chunk)
    
    wb.save('c:/myPycode/melona.xlsx')
    imageFile.close()
    print(imageFile)
```
