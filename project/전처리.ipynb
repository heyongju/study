```python
import pandas as pd

batter1_2020 = pd.read_excel('C:/users/PlayData/desktop/kbodata18_20.xlsx', sheet_name = 13 )#wrc
batter2_2020 = pd.read_excel('C:/users/PlayData/desktop/kbodata18_20.xlsx', sheet_name = 14)#clutch
batter3_2020 = pd.read_excel('C:/users/PlayData/desktop/kbodata18_20.xlsx', sheet_name = 15 )#raa주루
batter4_2020 = pd.read_excel('C:/users/PlayData/desktop/kbodata18_20.xlsx', sheet_name = 16 )#raa수비
batter5_2020 = pd.read_excel('C:/users/PlayData/desktop/kbodata18_20.xlsx', sheet_name = 16 )#출장타석
batter6_2020 = pd.read_excel('C:/users/PlayData/desktop/kbodata18_20.xlsx', sheet_name = 17 )#장타율
```

```python
batter_2018_wrc=batter1_2020[['Unnamed: 2', 'Unnamed: 4']].iloc[2:282]
batter_2018_clutch = batter2_2020[['Unnamed: 2', 'Unnamed: 8']].iloc[2:282]
batter_2018_run = batter3_2020[['Unnamed: 2', 'Unnamed: 9']].iloc[2:282]
batter_2018_def = batter4_2020[['Unnamed: 2', 'Unnamed: 12']].iloc[2:282]
batter_2018_box = batter5_2020[['Unnamed: 2', 'Unnamed: 5']].iloc[2:282]
batter_2018_pow = batter6_2020[['Unnamed: 2', 'Unnamed: 26']].iloc[2:282]
batter_2018_wrc
```

```python
batter_2018_wrc.rename(columns = {'Unnamed: 2':'name', 'Unnamed: 4':'wrc+'},inplace = True)
batter_2018_clutch.rename(columns = {'Unnamed: 2':'name', 'Unnamed: 8':'득점권타율'},inplace = True)
batter_2018_run.rename(columns = {'Unnamed: 2':'name', 'Unnamed: 9':'RAA주루'},inplace = True)
batter_2018_def.rename(columns = {'Unnamed: 2':'name', 'Unnamed: 12':'수비RAA종합'},inplace = True)
batter_2018_box.rename(columns = {'Unnamed: 2':'name', 'Unnamed: 5':'출장타석'},inplace = True)
batter_2018_pow.rename(columns = {'Unnamed: 2':'name', 'Unnamed: 26':'장타율'},inplace = True)
```

```python
excel_writer = pd.ExcelWriter('C:\mypycode/data/2018타자(wrc).xlsx', engine = 'xlsxwriter')
batter_2018_wrc.to_excel(excel_writer, index = False)
excel_writer.save()
```

```python
batter_2018_middle1 = pd.merge(batter_2018_wrc, batter_2018_clutch, left_on='name', right_on='name', how='outer' )
batter_2018_middle2 = pd.merge(batter_2018_middle1, batter_2018_run, left_on='name', right_on='name', how='outer' )
batter_2018_middle3 = pd.merge(batter_2018_middle2, batter_2018_def, left_on='name', right_on='name', how='outer' )
batter_2018_middle4 = pd.merge(batter_2018_middle3, batter_2018_pow, left_on='name', right_on='name', how='outer' )
```
