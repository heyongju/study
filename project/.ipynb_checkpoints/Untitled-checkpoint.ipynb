```python
import pandas as pd
import datetime
import urllib
import openpyxl
# print(urllib)

read = pd.read_excel('C:/Users/Playdata/Desktop/kbo_no_salary.xlsx')

    
att = read[['포수', '생년월일']]
# print(read.__dict__)
# print(len(read['포수']))

for i in range(0, len(read['포수'])):
    name = read['포수'][i]
    birth = read['생년월일'][i].date()
    print(f'name: {name} birth: {birth}')
    excel_writer = pd.ExcelWriter('C:\mypycode/data/111.xlsx', engine = 'xlsxwriter')
    get_salary(name, birth)


#     df[0].to_excel(excel_writer, index = False)


def get_salary(name, birth): 
    # pandas.read_html
    # 지정한 홈페이지 주소에 있는 테이블()들을 DataFrame의 List로 반환
    wb = openpyxl.Workbook()
    sheet = wb.active
    sheet.append(['연도', '연봉', 'WAR'])
    name = urllib.parse.quote(name)
#     print(name)
    url = f'http://www.statiz.co.kr/player.php?opt=11&name={name}&birth={birth}'

    attrs = {
           'class': 'table-condensed'
    }
    df = pd.read_html(url, header = 0, attrs=attrs)
    sheet.append(df[0])
#     print(df[0])

excel_writer.save()

"""
1. 기존 csv
2. for 문으로 기존 csv
3. 데이터가져오는거
4. pd.dataframe 넣고
5. for문 끝나면 저장
"""
```
