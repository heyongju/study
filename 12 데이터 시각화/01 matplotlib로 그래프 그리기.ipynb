```python
import numpy as np

x = np.arange(-4.5, 5, 0.5) #배열 x 생성. 범위: [-4.5, 5], 0.5씩 증가
y = 2*x**2 #수식을 이용해 배열 x에 대응하는 배열 y 생성
print([x,y])
plt.plot(x,y)
plt.show()
```

#### 여러 그래프 그리기

```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
plt.plot(x, y1)
plt.figure()       #새 그래프 창을 생성
plt.plot(x, y2)
plt.show()
```

```python
# 데이터 생성

x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2  #NumPy에서 cos()는 np.cos()으로 입력

plt.figure(1)  #1번 그래프 창 생성함
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 글미

plt.figure(2)  #2번 그래프 창을 생성함
plt.plot(x, y2) #지정된 그래프 창에 그래프를 그림

plt.figure(1) #이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2) #지정된 그래프 창에 그래프를 그림

plt.figure(2) #이미 생성된 2번 그래프 창을 지정함
plt.clf() #2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1) #지정된 그래프 창에 그래프를 그림

plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2 #numPy에서 sin()은 np.sin()으로 입력
y4 = 10*np.exp(-x) +1 #NumPy에서 exp()는 np.exp()로 입력

# 2 x x 행렬로 이뤄진 하위 그래프에서 p에 따라 위치를 지정

plt.subplot(2,2,1) #p는 2행 2열 보기에서 1번
plt.plot(x,y1)

plt.subplot(2,2,2) #p는 2행 2열 보기에서 2번
plt.plot(x, y2)

plt.subplot(2,2,3) # p는 3
plt.plot(x, y3)

plt.subplot(2,2,4) #p는 4
plt.plot(x, y4)
plt.show()
```

### 그래프 꾸미기

#### 출력 형식 지정

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x,y1,'m', x,y2,'y', x,y3,'k', x,y4,'c')  #색 지정
plt.show()
```

```python
plt.plot(x,y1,'-', x,y2,'--', x,y3,':', x,y4,'-.')  #선 스타일 지정
plt.show()
```

#### 라벨, 제목, 격자, 범례, 문자열 표시

```python
x = np.arange(-4.5, 5, 0.5)
y = 2*x**3

plt.plot(x,y,'g--')
plt.xlabel('X-axis') #x축 이름
plt.ylabel('Y-axis') #y축 이름
plt.title('Graph title') #표 이름
plt.show() 
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x,y1,'>--r', x,y2,'s-y', x,y3,'d:b', x,y4,'-.Xc')  #색 지정
plt.grid() #격자생성
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x,y1,'>--r', x,y2,'s-y', x,y3,'d:b', x,y4,'-.Xc')  #색 지정
plt.legend(['data1', 'data2', 'data3', 'data4'], loc = 4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
plt.grid() 
```

### 산점도

```python
import matplotlib.pyplot as plt

height = [165, 177, 160, 180, 185, 155, 172]  #키 데이터
weight = [62, 67, 55, 74, 90, 43, 64]  #몸무게 데이터

size = 100 * np.arange(1,8) #데이터별로 마커의 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']  #데이터별로 마커의 컬러 지정
plt.scatter(height, weight, s=size, c=colors)  # s는 점 크기, c는 색
plt.xlabel('Height(m)')
plt.ylabel('Weight(Kg)')
plt.title('Height & Weight')
plt.grid()
```

```python
import numpy as np

city = ['서울', '인천', '대전', '대구', '울산', '부산', '광중']

# 위도(latitude)와 경도(longitude)
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

#인구 밀도(명.km^2): 2017년 통계청 자료
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2 #마커의 크기 지정(인구 밀도 별로)
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']  #마커의 컬러 지정

plt.scatter(lon, lat, s= size, c = colors, alpha = 0.5) #alpha = 마커의 투명도
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

for x,y,name in zip(lon, lat, city):
    plt.text(x,y,name) #위도, 경도 위치에 city 값 배치
    
plt.grid(False)
```

### 막대그래프

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04'] #회원 ID
before_ex = [27, 35, 40, 33]  # 운동 시작 전
after_ex = [30, 38, 42, 37]  #운동 시작 후

import matplotlib.pyplot as plot
import numpy as np

n_data = len(member_IDs)   #회원이 네 명이므로 전체 데이터 수는 4
index = np.arange(n_data)  #numpy를 이용해 배열 생성 (0, 1, 2, 3)
colors = ['r', 'g', 'b', 'm']


plt.figure(1)
plt.bar(index, before_ex, tick_label = member_IDs, color = colors, width = 0.6)
plt.grid(False)
plt.show()
#bar(x,y)에서 x = index, height = before_rx로 지정, x축 이름을 회원_ID로 설정
#막대 색상 설정, 막대 넓이는 0.6으로


plt.figure(2)
plt.barh(index, before_ex, tick_label = member_IDs, color = colors)  
#barh는 width를 사용할 수 없음 
plt.grid(False)
```

```python
barwidth = 0.4
plt.bar(index, before_ex, color = 'c', align = 'edge', width = barwidth, label = 'before')
plt.bar(index + barwidth, after_ex, color = 'm', align = 'edge', width = barwidth, label = 'after')

plt.xticks(index + barwidth, member_IDs)
plt.legend()
plt.xlabel("회원 ID")
plt.ylabel("윗몸일으키기 횟수")
plt.title('운동 시작 전과 후의 근지구력(복근) 변화 비교')
plt.show()
```

### 파이 그래프

```python
fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
result = [7, 6, 3, 2, 2]

import matplotlib.pyplot as plt

plt.pie(result, labels = fruit, autopct = '%.1f%%', startangle = 90, counterclock = False)
# 레이블은 fruit어레이에서 가져오고, 
# auotpct = 오토 퍼센트, 소숫점 1까지만 표시
# startangle은 원점기준 x축 각도서부터 시작.
# counterclock은 시계 반대 반향으로 어레이 순서배열함. False하면 시계방향으로 표시
plt.show()
```

```python
explode_value = (0.1, 0, 0, 0, 0)

plt.pie(result, labels=  fruit, autopct='%.1f%%', startangle = 90, counterclock = False,
       explode = explode_value, shadow = True)
plt.show()
```
